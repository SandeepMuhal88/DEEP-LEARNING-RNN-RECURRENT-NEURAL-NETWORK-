Goal: Predict the next timestep given a sequence.

Skills: Many-to-many, return_sequences=False for last layer.

```python
import numpy as np 
import tensorflow 
from tensorflow import keras
from keras.layers import SimpleRNN, Dense
from keras.models import Sequential
```

```python
seq_len=20
feature=5
```

```python
X=np.random.rand(1000,seq_len,feature).astype(np.float32)
```

```python
X[1]
```

```python
np.sum(X[:,-1,:][1])+0.5
```

```python
y=np.sum(X[:,-1,:],axis=1,keepdims=True)+0.5
```

```python
y
```

```python
model=Sequential()
model.add(SimpleRNN(20,return_sequences=True,input_shape=(seq_len,feature)))
model.add(SimpleRNN(10,return_sequences=True))
model.add(SimpleRNN(5,return_sequences=True))
model.add(SimpleRNN(2))
model.add(Dense(1))
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X, y, epochs=100, batch_size=32, verbose=1)
```

```python
X[:1][0][0]
```

```python
print("Pred:", model.predict(X[:1])[0][0])
```
