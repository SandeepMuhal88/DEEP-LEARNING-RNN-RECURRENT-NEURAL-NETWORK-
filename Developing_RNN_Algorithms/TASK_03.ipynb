#### END TO END MODEL IN REcurrent Neural Network

```python
import numpy as np
import tensorflow 
from tensorflow import keras
from keras.models import Sequential
from keras.layers import SimpleRNN,Dense
```

```python
num_samples = 500
seq_len = 5
features = 3
```

```python
X = np.random.rand(num_samples, seq_len, features).astype(np.float32)
y = np.sum(X[:, -1, :], axis=1, keepdims=True)  # sum of last timestep
```

```python
X
```

```python
y
```

```python
model= Sequential()
model.add(SimpleRNN(5,return_sequences=True,input_shape=(6,3)))
model.add(SimpleRNN(4,return_sequences=True))
model.add(SimpleRNN(2))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
history=model.fit(X,y,epochs=20,batch_size=16,verbose=1)
```

```python
test_sample = np.random.rand(1, seq_len, features).astype(np.float32)
pred = model.predict(test_sample)
```

```python
print("\nTest sample:\n", test_sample)
print("True sum of last timestep:", np.sum(test_sample[:, -1, :]))
print("Model prediction:", pred[0][0])
```
